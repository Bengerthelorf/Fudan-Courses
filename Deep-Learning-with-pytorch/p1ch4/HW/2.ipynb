```python
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, linewidth=75)
```

```python
with open('./text.txt', encoding='utf8') as f:
    text = f.read()
```

```python
line = text.split('\n')
len(line), line[21]
```

```python
def clean_words(input_str):
    punctuation = '.,;:"!?”“_-()'
    word_list = input_str.lower().replace('\n',' ').split()
    # word_list = input_str.replace('-',' ').split()
    # word_list = input_str.replace(',',' ').split()
    word_list = [word.strip(punctuation) for word in word_list]
    return word_list
```

```python
words_list = sorted(set(clean_words(text)))
words_list
```

```python
word2index_dict = {word: i for (i, word) in enumerate(words_list)}
```

```python
text_in_words = clean_words(text)
```

```python
text_one_hot = torch.zeros(len(text_in_words), len(words_list))
```

```python
for i, word in enumerate(text_in_words):
    text_one_hot[i][word2index_dict[word]] = 1

print("Text one hot encoding: \n", text_one_hot)
print("Size of one hot encoding: \n", text_one_hot.size())
```
